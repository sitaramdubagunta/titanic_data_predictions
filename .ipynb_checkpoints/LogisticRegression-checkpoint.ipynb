## Feature engineering the ticket feature

```python

 titanic.groupby('Ticket')['Ticket'].count().sort_values(ascending  = False)
```

####  here we can see many people have 1 single ticket this might be useful for our model


```python
titanic['count_ticket']  =  titanic.groupby('Ticket')['Ticket'].transform('count')
titanic
```

####  the reason why a few people share the same TICKET is because they might be part of the same group or family

```python
titanic.Ticket.apply(lambda x :  'PC' if x.startswith('PC') else 'Other' ).value_counts()
```

### It is intresting to note that there are many people with tickets starting with PC and W.C/ and so on but for now we wont group it based on that instead we will be dropping the tickets coloumn


```python
titanic = titanic.drop(['Ticket'] , axis = 1)
```

# Encoding the categorical data (one hot encoding)

#### we will be using one hot encoding over label encoding since in label encoding for example in EMBARKED (S,C,Q) label encoding stores it as (1,2,3) 
#### which might give a bias towards Q if we use label encoding


```python
titanic = pd.get_dummies(titanic , drop_first = True)
```

```python
titanic
```

# Feature scaling using standarization

```python
feature = titanic.iloc[ :, 1:]
target = titanic.iloc[: ,0]
feature_toscale  = ['Age' , 'Fare']

for column in feature_toscale:

    titanic[column] = (titanic[column] - titanic[column].mean() ) / titanic[column].std()
```

```python
titanic
```

```python
for i , column in enumerate(titanic , 1):
    plt.subplot(4,4,i)
    sns.boxplot(titanic[column])
plt.tight_layout()
```
